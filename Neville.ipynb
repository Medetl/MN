```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([0 ,1 ,2 ,3 ,4 ,5])
y=np.array([1 ,1 ,2 ,6 ,24 ,120])
```

```python
n=x.size
G=np.zeros((n,n))
xi=np.pi
yi =0
```

```python
G[:,0]= y
for i in range(1,n):
    for j in range(1,i +1) :
        G[i,j ]=(( xi -x[i-j])*G[i,j -1] -(xi -x[i])*G[i -1,j -1]) /(x[i]-x[i-j])
        yi=G[i,i]
        if abs(G[i,i]-G[i -1,i -1]) <0.00001:
            exit
```

```python
print("\nTabla G:")
encabezado = "i      xi             " + "       ".join([f"G(i, {j})" for j in range(n)])  #Crea una comprensión de lista que crea una lista de cadenas
print(encabezado)

for i in range(n):
    fila = [f"{G[i, j]:.7f}" if i >= j else "         " for j in range(n)]
    print(f"{i:<6}{x[i]:<14.7f}" + "   ".join(fila)) #join une todos los elementos de la lista, # Imprime cada fila:i, x[i] y Q[i,j]

# El valor del polinomio evaluado en xp
Pxp = G[n-1, n-1]
print(f"\nP({yi}) = {Pxp:.7f}") #Aqui en el .7 se puede cambiar dependiendo cuantos decimales quiere.
```

```python
plt.plot(x,y,'o')
plt.plot(xi,yi,'sr')
plt.text(xi +0.1,yi,'$\pi$!' + str(yi))
plt.title('Factorial n!')
plt.legend(['Factorial','Interpolación'])
plt.grid( True )
plt.show()
```
